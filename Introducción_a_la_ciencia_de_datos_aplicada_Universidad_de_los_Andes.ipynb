#### Tablas de contingencia:

```python
## Librerias adicionales para la prueba chi-cuadrado con las tablas de contingencia:
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
productos_top = datos.groupby('nombre_categoria_producto').count().reset_index()[['nombre_categoria_producto', 'producto_id']]
cat = productos_top.head()['nombre_categoria_producto']
cat.head()
```

```python
productos_top1 = datos.groupby('departamento_vendedor').count().reset_index()[['departamento_vendedor', 'producto_id']]
departamento_top_lista = productos_top1.head()['departamento_vendedor'].to_list()
departamento_top_lista
```

```python
## Filtrar por departamento top:
datos_top = datos[datos['nombre_categoria_producto'].isin(cat) & datos['departamento_vendedor'].isin(departamento_top_lista)]
datos_top.head()
```

```python
## Realizar la tabla de contingencia:
tabla_cont = pd.crosstab(datos_top['nombre_categoria_producto'], datos['departamento_vendedor'], margins=True)
tabla_cont
```

```python
## Análisis fila:
pd.crosstab(datos_top['nombre_categoria_producto'], datos['departamento_vendedor'], margins=True, normalize = 'index')
```

```python
## Análisis columna:
pd.crosstab(datos_top['nombre_categoria_producto'], datos['departamento_vendedor'], margins=True, normalize = 'columns')
```

```python
## Análisis totales:
pd.crosstab(datos_top['nombre_categoria_producto'], datos['departamento_vendedor'], margins=True, normalize = 'all')
```

```python
tabla_cont
```

```python
## Utilizar el mínimo para columnas:
tabla_cont.min()
```

```python
## Utilizar el mínimo para filas:
tabla_cont.transpose().min()
```

```python
## Prueba Chi-cuadrado:
chi, pval, dof, exp = chi2_contingency(tabla_cont)
print("El valor p es: ", pval)
print("El valor del Estadística de Prueba es: ", chi)
print("Los grados de libertad son: ", dof)

valor_critico = chi2.ppf(0.95, dof)
print("El valor critico de la distribución chi-cuadrado es: ", valor_critico)
print("Por lo que en este caso no se rechaza la hipótesis nula H0 ya que el valor chi no es mayor al valor critico")
print("Así, concluimos que las variables son independientes")
```

## **Semana 4: Comparación entre grupos y validación de modelos estadísticos**

#### Intervalos de confianza y el p-test en Python

```python
## Importemos primero las librerias necesarias:
import pandas as pd
# Para manejo de datos y calculos
import numpy as np
# Calcular el test estadístico:
from statsmodels.stats import weightstats as stests
# Libreria para graficar:
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("/content/Ordenes_productos_C1_M2.csv", sep = ";", encoding = "Latin-1")
datos.head(5)
```

```python
## Saquemos el I.C para el valor de flete medio:
x = datos['valor_flete'].mean()
alpha = 0.05
sigma = datos['valor_flete'].std()
n = len(datos['valor_flete'])
z = 1.96 # Con alpha = 0.05
```

```python
# Apliquemos la formula, primero saquemos la margen de error del intervalo:
# dado por:
error = z*(sigma/np.sqrt(n))
intervalo = [x-error, x+error]
intervalo
```

```python
# Ahora, vamos a realizar una prueba de hipótesis para la diferencia de medias de
# de dos departamentos: Antioquia vs. Bogotá d.c
datos_flete = datos.groupby('departamento_vendedor').mean().reset_index()[['departamento_vendedor', 'valor_flete']]
# Ahora, grafiquemos y notemos el comportamiento:
datos_flete.plot.bar(x = 'departamento_vendedor', y = 'valor_flete')
```

```python
# La hipótesis serían, H0 = El promedio de los dos departamentos son iguales vs
# H1: El promedio de los dos departamentos son diferentes
flete_antioquia = datos[datos['departamento_vendedor']=="Antioquia"]['valor_flete']
flete_bogota = datos[datos['departamento_vendedor']=="Bogota d.c."]['valor_flete']
# Imprimamos el valor de X_barra para casa uno de los departamentos:
print("Media de Antioquia: ", flete_antioquia.mean())
print("Media de Bogotá d.c: ", flete_bogota.mean())
```

```python
# Valores del E.P z y del valor p:
ztest, pvalor = stests.ztest(flete_antioquia, x2 = flete_bogota, value = 0, alternative="two-sided")
print("El E.P Z es: ", ztest)
print("El valor p de la prueba es: ", pvalor)
```

Con un nivel de significacia de alpha = 0.05 concluimos que no se rechaza la hipótesis nula H0 y se concluye que el promedio de los dos departamentos son iguales.

#### Análisis de varianza - ANOVA:
Para pobrar diferencia de medias en más de dos poblaciones. Bajo la hipótesis nula de que todas las medias son iguales y la alternativa de que hay al menos una diferente.

```python
## Carguemos librerias necesarias:
import pandas as pd
import scipy.stats as stats ## Para calcular el p valor de la F

# NOTA: los datos ya están cargados del ejercicio anterior
```

```python
## Busquemos la categoria top de productos de nuevo:
categoria_top=datos.groupby('nombre_categoria_producto').count().reset_index()[['nombre_categoria_producto','producto_id']].sort_values(by=['producto_id'], ascending=False)
categoria_top_lista=categoria_top.head(1)['nombre_categoria_producto']
categoria_top.head(1)
```

```python
## Encontremos los departamentos top 5:
departamento_top=datos.groupby('departamento_vendedor').count().reset_index()[['departamento_vendedor','producto_id']].sort_values(by=['producto_id'], ascending=False)
departamento_top_lista=departamento_top.head()['departamento_vendedor'].to_list()
departamento_top.head()
```

```python
## Filtrar por los departamentos top:
datos_top=datos[datos['nombre_categoria_producto'].isin(categoria_top_lista) & datos['departamento_vendedor'].isin(departamento_top_lista)]
datos_top[['precio','departamento_vendedor','nombre_categoria_producto']].head()
```

```python
# 1. Calcular el tamaño de la muestra de cada categoria
datos_muestra = datos_top[['precio', 'departamento_vendedor']]
tamano_grupos = datos_muestra.groupby('departamento_vendedor').count().reset_index()
tamano_grupos

## El reset_index() es una función para enumerar la lista de abajo hacia arriba
```

```python
tamano_grupos.plot(kind = 'bar', x = 'departamento_vendedor', y = 'precio')
```

```python
## 2. Hallar las varianzas para los grupos:
varianza_grupos = datos_muestra.groupby('departamento_vendedor').std()
varianza_grupos
```

Ya con estos dos elementos podemos proceder al cálculo de la tabla, con esto podremos realizar la prueba de hipótesis y mirar si existen diferencias entre los promedios de los precios en las 5 ciudades que más se venden:
SST representa la suma de varianzas de cada dato con respecto a la varianza total
SSR o SSA representa la suma de varianzas con respecto al factor A
SSE representa la suma de varianzas que no pueden ser explicadas.

```python
# Crear la estructura de la tabla ANOVA
data = [['SSR', '', '', '', '', ''], ['SSE', '', '', '', '', ''], ['Total', '', '', '', '', '']]
anova_tabla = pd.DataFrame(data, columns = ['Fuente de variación', 'SS', 'gdl', 'MS', 'F', 'P-valor'])
anova_tabla.set_index('Fuente de variación', inplace = True)
anova_tabla
```

```python
# Como ya creamos la tabla, vamos a rellenarla:
# Calculamos SSR y actualizamos la tabla
promedio_precio = datos['precio'].mean()
SSR = datos_muestra.groupby('departamento_vendedor').count() * (datos_muestra.groupby('departamento_vendedor').mean() - promedio_precio)**2
anova_tabla.at['SSR', 'SS'] = SSR['precio'].sum()
anova_tabla.at['SSR', 'SS']
```

```python
# Calculamos la SSE y actualizar en la tabla
SSE = (datos_muestra.groupby('departamento_vendedor').count() - 1) * datos_muestra.groupby('departamento_vendedor').std()**2
anova_tabla.at['SSE', 'SS'] = SSE['precio'].sum()
anova_tabla.at['SSE', 'SS']
```

```python
# Calculamos SST y actualizar en la tabla
SST = SSR['precio'].sum() + SSE['precio'].sum()
anova_tabla.at['Total', 'SS']= SST
anova_tabla.at['Total', 'SS']
```

```python
# Calculamos los grados de libertas
anova_tabla['gdl']['SSR'] = datos_muestra['departamento_vendedor'].nunique() - 1
anova_tabla['gdl']['SSE'] = datos_muestra.shape[0] - datos_muestra['departamento_vendedor'].nunique()
anova_tabla['gdl']['Total'] = datos_muestra.shape[0] - 1
anova_tabla['gdl']['SSR']
```

```python
anova_tabla['gdl']['SSE']
```

```python
anova_tabla['gdl']['Total']
```

```python
# Calculamos MS
anova_tabla['MS'] = anova_tabla['SS'] / anova_tabla['gdl']
anova_tabla['MS']
```

```python
# Calculamos F
F = anova_tabla['MS']['SSR'] / anova_tabla['MS']['SSE']
anova_tabla.at['SSR','F'] = F
F
```

```python
# Calculamos el p-valor
anova_tabla.at['SSR','P-valor'] = 1 - stats.f.cdf(F, anova_tabla.at['SSR','gdl'], anova_tabla.at['SSE','gdl'])
anova_tabla.at['SSR','P-valor']
```

```python
# Visualizando la tabla ANOVA:
anova_tabla
```

```python
# Evaluar la hipótesis con un nivel de confianza del 95% y significancia del 5%
alpha=0.05 #Establecer la significancia
# Evaluar la hipótesis
print("Usar el P-valor para decidir sobre la hipótesis")
if anova_tabla.at['SSR','P-valor'] <= alpha:
    conclusion = "Se rechaza hipótesis nula "
else:
    conclusion = "No se rechaza hipótesis nula "
print(conclusion)
print("Valor F es:", anova_tabla['F']['SSR'], " y el P-valor es:", anova_tabla['P-valor']['SSR'])
```

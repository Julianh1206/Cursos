## **Semana 2:**
*Intervalos de confianza*

Ejercicios de evaluación de I.C

```python
import numpy as np
import pandas as pd
clientes = pd.read_excel('/content/Datos_fugaclientes.xlsx')
clientes.head(5)
```

```python
## Ej 1: I.C al 95% para la media de la duración media de llamadas al mismo operador (Suponiendo que es normal)
X_barra1 = clientes['Duración promedio llamadas al mismo operador'].mean()
S1 = clientes['Duración promedio llamadas al mismo operador'].std()
n1 = len(clientes['Duración promedio llamadas al mismo operador'])
n1 = 291
cuantil_t = 1.968178 ## Grados de libertad es n-1 = 290
margen_error1 = cuantil_t*(S1/np.sqrt(n1))
intervalo1 = [X_barra1-margen_error1, X_barra1+margen_error1]
intervalo1
```

```python
## Ej 2: I.C al 95% para la varianza de la duración promedio de llamadas a otros operadores (Suponiendo que es normal)
# Lo único que necesitamos es el tamaño de la muestra en este vector, la varianza muestral y los dos cuantiles chi-sq
n2 = len(clientes['Duración promedio llamadas a otros operadores'])
n2 = 291
varianza_otro = clientes['Duración promedio llamadas a otros operadores'].var()
chisq_bajo = 339.067
chisq_alto = 244.7197
intervalo2 = [((n2-1)*varianza_otro)/chisq_bajo, ((n2-1)*varianza_otro)/chisq_alto]
intervalo2
```

```python
## Ej 3: I.C al 90% para el cociente de varianzas entre la duración promedio de llamadas al mismo
## operador y la duración promedio de llamadas a otros operadores:

# varianza para el mismo operador:
S1 = S1*S1

# Varianza para otro perador:
varianza_otro

# Cuantiles:

f_bajo = 1.213484
f_alto = 0.8240737

intervalo3 = [(S1/varianza_otro)*f_bajo, (S1/varianza_otro)*f_alto]
intervalo3.sort()
intervalo3
```

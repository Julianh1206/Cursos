```python
# Equivalent Value and Frequency

#(para números pequeños)

nums = input()

def solve(nums):
  lista = list(nums)
  lista = list(map(int,lista)) ## map: convierto elemento por elemento en lo que yo quiera
  lista1 = set(lista.copy()) ## set: pone elementos repetidos solo una vez
  for i in lista1:
    if lista.count(i) == i:
        return True
  return False

print(solve(nums))
```

## Catedra Pedro Nel

```python
### IDEA PRINCIPAL

def funcion(mu, x0, n):
  lista = []
  contador = 0
  x1 = mu*x0*(1-x0)
  lista.append(x1)
  contador += 1
  if contador <= n:
    funcion(mu, x1, n-1)
  return lista

funcion(2.8, 0.2, 100)
```

```python
## Verisón mejorada con un for

## Función recursiva para iteraciones:
def funcion(mu, x0, n):
    lista = []
    for _ in range(n):
        x1 = mu * x0 * (1 - x0)
        lista.append(x1)
        x0 = x1
    return lista

## Otra forma:
def iteracion(u, inicio, iteraciones):
    return [inicio := u * inicio * (1 - inicio) for _ in range(iteraciones)]
```

```python
## literal a) Escogiendo a mu = 3.1 y punto inicial x0 = 0.2 para un período igual a 1
resultados = funcion(2.9, 0.2, 100)
print(resultados)

## Gráfica:
import matplotlib.pyplot as plt
# Graficar los resultados
plt.plot(resultados, 'b-', markersize=2)
plt.xlabel('Iteraciones')
plt.ylabel('Valores de x')
plt.title('Diagrama del Mapa Logístico')
plt.grid(True)
plt.show()
```
